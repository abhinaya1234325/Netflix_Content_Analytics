```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("netflix_titles.csv")
print(df.shape)
print(df.columns)
print(df.dtypes)
df.head()
```

```python
df.isnull().sum()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
print(df)
```

```python
df['year_added'] = df['date_added'].dt.year
```

```python
print(df)
```

```python
sns.countplot(data=df, x='type', palette='pastel')
plt.title("Count of Movies vs TV Shows")
plt.show()
```

```python
df['year_added'].value_counts().sort_index().plot(kind='bar', figsize=(12,6), color='skyblue')
plt.title("Content Added by Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.grid(True)
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)

top_countries.plot(kind='bar', color='coral')
plt.title("Top 10 Countries Producing Netflix Content")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("netflix_titles.csv")
```

```python
print(df.columns)
```

```python
df['listed_in'].value_counts().head(10).plot(kind='barh', color='teal')
plt.title("Top Genres on Netflix")
plt.xlabel("Count")
plt.ylabel("Type")
plt.gca().invert_yaxis()
plt.show()
```

```python
sns.countplot(data=df, y='rating', order=df['rating'].value_counts().index, palette='mako')
plt.title("Content Ratings on Netflix")
plt.show()
```
